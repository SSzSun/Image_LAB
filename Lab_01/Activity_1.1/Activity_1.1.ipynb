```python
# import Lib
import cv2
import numpy as np
import math
import matplotlib.image as mping

from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
img_1 = mping.imread("fish.jpg")
img_2 = mping.imread("girl.jpg")
```

```python
BGR = cv2.cvtColor(img_1, cv2.COLOR_BGR2RGB)
xB = BGR[:,:,0]
xG = BGR[:,:,1]
xR = BGR[:,:,2]

yR = img_1[:,:,0]
yG = img_1[:,:,1]
yB = img_1[:,:,2]

# Config Show Picture
# Line 1
plt.subplot(2,4,1);plt.imshow(BGR);plt.title("BGR")
plt.subplot(2,4,2);plt.imshow(xB, cmap='gray');plt.title("B")
plt.subplot(2,4,3);plt.imshow(xG, cmap='gray');plt.title("G")
plt.subplot(2,4,4);plt.imshow(xR, cmap='gray');plt.title("R")

# Line 2
plt.subplot(2,4,5);plt.imshow(img_1);plt.title("RGB")
plt.subplot(2,4,6);plt.imshow(yR, cmap='gray');plt.title("R")
plt.subplot(2,4,7);plt.imshow(yG, cmap='gray');plt.title("G")
plt.subplot(2,4,8);plt.imshow(yB, cmap='gray');plt.title("B")
```

```python
X = np.transpose(img_2)
Y = np.moveaxis(img_2, 2, 0)
height, width, channels = img_2.shape
R = np.reshape(img_2, (channels, height, width))

print('Image Shape : ',img_2.shape)
print('Tranpose --> ',X.shape)
print('Moveaxis --> ',Y.shape)
print('Reshape --> ',R.shape)

plt.subplot(2, 2, 1);plt.imshow(img_2[:,:, 0], cmap='gray');plt.title("Image_Original")
plt.subplot(2, 2, 2);plt.imshow(X[0, :, :], cmap='gray');plt.title("Image_Tranpose")
plt.subplot(2, 2, 3);plt.imshow(Y[0, :, :], cmap = "gray");plt.title("Image_MoveAxis")
plt.subplot(2, 2, 4);plt.imshow(R[0, :, :], cmap= "gray");plt.title("Image_Reshape")
```

```python
image = cv2.cvtColor(img_2, cv2.COLOR_RGB2GRAY)

# print("Original",image)
plt.subplot(1, 2, 1);plt.imshow(image,cmap='gray');plt.title('Original')

def BitQuantize(img):
    H,W = img.shape
    for w in range(0,W):
        for h in range(0,H):
            img[h][w] = (img[h][w]/255)*15
    return img

img_Bit = BitQuantize(image)
plt.subplot(1, 2, 2);plt.imshow(img_Bit,cmap='gray');plt.title('Quantize')
# print("Quantize",img_Bit)
```

```python
img_gray = cv2.cvtColor(img_2, cv2.COLOR_BGR2GRAY)
img_resize = cv2.resize(img_gray, (300, 300))

plt.imshow(img_resize, cmap='gray')

# create the x and y coordinate arrays
x, y = np.mgrid[0:img_resize.shape[0], 0:img_resize.shape[1]]

# create the figure
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.plot_surface(x, y, img_resize, cmap=plt.cm.gray)
ax.set_zlim(0,256)
ax.zaxis.set_major_formatter('{x:.02f}')
```
